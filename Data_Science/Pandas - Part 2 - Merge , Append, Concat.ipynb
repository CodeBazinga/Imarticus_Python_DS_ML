```python
import pandas as pd
import numpy as np
```

```python
# dataFrame 1
d1 = {'Customer_id':[1,2,3,4,5,6],
     'Product':['Oven','Oven','Oven','Television','Television','Television']}
df1 = pd.DataFrame(d1)
 
# data frame 2
d2 = {'Customer_id':[2,4,6,7,8],
    'State':['California','California','Texas','Austria','Spain']}
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

### Merge

<b>inner : </b> Return only the rows in which the left table has matching keys in the right table

```python
# Inner  
d1 = pd.merge(df1, df2, on='Customer_id', how='inner')
# d1[['Product','State']]
d1
```

<b>outer :</b> Returns all rows from both tables, join records from the left which have matching keys in the right table.

```python
pd.merge(df1, df2, on='Customer_id', how='outer')
```

<b>Left : </b> Return all rows from the left table, and rows with matching keys from the right table.

```python
pd.merge(df1, df2, on='Customer_id', how='left')
```

<b>Right :</b> Return all rows from the right table, and rows with matching keys from the left table.

```python
pd.merge(df1, df2, on='Customer_id', how='right')
```

```python
pd.merge(df1, df2, how='cross')
```

### Using merge with more than one columns in common

```python
# dataFrame 1
da1 = {'Customer_id':[1,2,3,4,5,6],
       'Name':['Ashish','Ridhima','Aanya','Akhilesh','Kunal','Shreya'],
      'Product':['Oven','Oven','Oven','Television','Television','Television']}
dfr1 = pd.DataFrame(da1)
 
# data frame 2
da2 = {'Customer_id':[2,4,6,7,8],
       'Name':['Ridhima','Akhilesh','Shreya','Watson','Vyshakh'],
       'State':['California','California','Texas','Austria','Spain']}
dfr2 = pd.DataFrame(da2)
```

```python
dfr1
```

```python
dfr2
```

#### 1) Inner merge based on two columns

```python
r1 = pd.merge(dfr1, dfr2, on=['Customer_id','Name'], how='inner')
r1
```

#### 2) Outer merge based on two columns

```python
pd.merge(dfr1, dfr2, on=['Customer_id','Name'], how='outer')
```

#### 3) Left merge based on two columns

```python
pd.merge(dfr1, dfr2, on=['Customer_id','Name'], how='left')
```

#### 4) Right merge baed on two columns

```python
pd.merge(dfr1, dfr2, on=['Customer_id','Name'], how='right')
```

### Concat
